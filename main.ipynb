```python
import regex as re

def remove_wiki_citation_glyphs(text: str) -> str:
    # Remove modifier letters, superscripts, phonetic extensions
    text = re.sub(r'\p{Lm}+', '', text)
    text = re.sub(r'\p{Sk}+', '', text)
    text = re.sub(r'\p{No}+', '', text)
    return text

def clean_wiki_text(text: str) -> str:
    # Remove [citation needed], [1], [2], etc.
    text = re.sub(r'\[[^\]]*\]', '', text)

    # Remove Unicode citation glyphs
    text = remove_wiki_citation_glyphs(text)

    return text.strip()

def remove_table_blocks(text: str) -> str:
    text = re.sub(r'(^|\n)(\|.*?)(?=\n[^|]|\Z)', '\n', text, flags=re.DOTALL)
    text = re.sub(r'\n{3,}', '\n\n', text)
    return text.strip()

def truncate_list_article(text: str) -> str:
    split_markers = [
        "\n## Lists",
        "\n== Lists",
        "\n### Lists"
    ]
    for marker in split_markers:
        if marker in text:
            return text.split(marker)[0]
    return text

def clean_text(text: str) -> str:  
    text = clean_wiki_text(text)
    text = remove_table_blocks(text)
    text = truncate_list_article(text)

    return text

clean = clean_text(text)
print(clean)
```

```python
print(tokenzier.decode(tokenzier.encode(clean)))
```
